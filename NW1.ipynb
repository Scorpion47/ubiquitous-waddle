```
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as pt
%matplotlib inline
x_data=np.linspace(0,10,10000)
noise=np.random.rand(len(x_data))
y_true=(0.5*x_data)+5+noise
batch_size=8
m=tf.Variable(0.309)
b=tf.Variable(0.737)
x_placeholder=tf.placeholder(dtype=tf.float32,shape=(batch_size))
y_placeholder=tf.placeholder(dtype=tf.float32,shape=(batch_size))
y_model=m*x_placeholder+b
error=tf.reduce_sum(tf.square(y_model-y_placeholder))
optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.001)
train=optimizer.minimize(error)
init= tf.global_variables_initializer()
loop=10000
with tf.Session() as sess:
    sess.run(init)
    for i in range(loop):
        rand_in1=np.random.randint(len(x_data),size=(batch_size))
        feed_dic={x_placeholder:x_data[rand_in1],y_placeholder:y_true[rand_in1]}
        sess.run(train,feed_dic)
        
    slope,intercept=sess.run([m,b])
  
y_new=(slope*x_data)+intercept     
```

```
featcols = [tf.feature_column.numeric_column('x',shape=[1])]
estimator=tf.estimator.LinearClassifier(feature_columns=featcols)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x_data,y_true,test_size=0.3,random_state=101)
input_fn1=tf.estimator.inputs.numpy_input_fn(x={'x':x_train},y=y_train,batch_size=8,num_epochs=None,shuffle=True)
estimator.train(input_fn1 ,steps=100)
```
